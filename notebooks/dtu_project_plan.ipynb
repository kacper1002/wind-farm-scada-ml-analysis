# Project 1: Wind farm SCADA dataset exploration

## Project summary
We are analyzing a NetCDF wind farm dataset from the Nørrekær Enge II wind farm (Denmark) provided by DTU Wind Energy. The dataset contains approximately 72k timestamps with 10-minute resolution and includes SCADA measurements from turbines and meteorological mast measurements.

The objective of this project is to understand and prepare the dataset before performing engineering or machine learning analysis.

**Data characteristics**
- Time series length: ~72,000 observations
- Sampling interval: 10 minutes
- Variables: ~260 signals

- Data sources:
    - turbine SCADA signals (power, nacelle wind speed, yaw etc.)
    - meteorological mast measurements
    - quality control flags (*_qc)

## Goals
1. Load and inspect the NetCDF dataset using xarray
2. Identify turbine signals (power, wind speed etc.)
3. Identify meteorological mast measurements
4. Inspect QC flags and missing values
5. Convert relevant variables to pandas DataFrame
6. Produce exploratory plots

## Expected outputs
- dataset structure overview
- mapping of turbine variables
- missing data statistics
- QC flag analysis
- basice time series plots

## Outcome
A cleaned and understood dataset ready for engineering analysis.


# Project 2: Wind turbine power curve analysis

## Project summary
Using cleaned dataset from Project 1, we will analyze the power curve of a wind turbine using SCADA data.

The goal is to study the relationship between wind speed and electrical power production, which is fundamental concept in wind turbine performance analysis.

## Engineering context
Wind turbines follow a characteristic power curve:
- cut-in region
- power ramp
- rated power region
- cut-out

SCADA data allow to reconstruct this curve using measured data

## Goals
1. Select turbine
2. Filter data using QC flags
3. Remove missing or invalid measurements
4. Plot scatter: wind speed vs power
5. Compute binned wind speed averages
6. Construct the empirical power curve
7. Compare scatter vs averaged curve

## Expected outputs
- Scatter plot 
- Binned power curve
- Interpretation of turbine behavior

## Outcome
A reconstructed power curve based on real SCADA data

# Project 3: Wind farm wake effect analysis

## Project summary
Using wind farm layout at Norrekaer Enge II, we will analyze wake effects between turbines.

Wake effects occur when upstream turbines reduce wind speed and increase turbulence for downstream turbines, using causing power losses.

## Dataset context
Wind farm contains 42 turbines arranged in grid (A1-F7) with spacing
- 6D within rows
- 7-8D between rows
Because of this layout, turbines aligned with wind direction can experience wake interactions

## Goals
1. Select wind direction sectors (e.g. 250-290)
2. Identify turbine rows aligned with wind
3. Compare power production across rows
4. Quantify wake losses downstream
5. Visualize wake effect using plots

## Expected outputs
- Wake loss plots
- Power comparison across turbine rows
- Wind-direction filtered analysis

## Outcome
Quantification of wake losses in the wind farm

# Project 5: Time series forecasting (LSTM)
## Project summary
This project applies deep learning (LSTM networks) to forecast turbine power using historical time series data.

Unlike Project 4 (which predicts instantaneous power) this project predicts future power output.

## Forecasting problem
Predict:
- Power (t+1)
- Power (t+6)
- Power (t+12)

## Techniques
- Sequence windowing
- Scaling
- Neural Networks (TensorFlow / Keras)

## Goals
1. Create time-windowed sequences
2. Normalize features
3. Train LSTM model
4. Evaluate forecast accuracy
5. Compare against baseline models

## Outcome 
A deep learning model for short term wind power forecasting
